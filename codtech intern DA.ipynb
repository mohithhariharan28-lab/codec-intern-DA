```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

df.head()
```

```python
print("Shape of data:", df.shape)
print("\nColumn names:")
print(df.columns)

print("\nSummary statistics:")
print(df.describe())
```

```python
sales = np.array(df['Sales'])

print("Mean:", np.mean(sales))
print("Max:", np.max(sales))
print("Min:", np.min(sales))
print("Sum:", np.sum(sales))

# Remove missing values
df = df.dropna()

# Check again
df.info()
plt.figure(figsize=(8,5))
plt.bar(df['Category'], df['Sales'])
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

```

```python

sns.histplot(df['Sales'], bins=10)
plt.title("Sales Distribution")
plt.show()
df.to_excel("cleaned_output.xlsx", index=False)
files.download("cleaned_output.xlsx")

```
